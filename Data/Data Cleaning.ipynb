```python
import pandas as pd
import numpy as np
df = pd.read_csv('~/Downloads/Hackathon Data.csv') 
df2 = df.drop(['Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13'], axis = 1)
df3 = df2.dropna(axis=0)
df3["SALES_DATE"] = df3["SALES_DATE"].astype('datetime64[ns]')
df3
```

```python
# Turn skus into rows and have units sold, promo price, retail price, competitor price, and inventory into columns
# Chaotic and ass
df4 = pd.pivot_table(df3, values=['DAILY_UNITS', 'RETAIL_PRICE', 'PROMO_PRICE', 'COMPETITOR_PRICE', 'Inventory'], index=['Encoded_SKU_ID', 'SUBCLASS_NAME', 'CLASS_NAME', 'ML_NAME', "CATEGORY_NAME"], columns=['SALES_DATE'], aggfunc = np.sum).fillna(0.0)
#df4.columns = df4.columns.categories

df4.reset_index()
df4
```

```python
# Turn rows into dates and skus and their aspects into columns
# Also ass
df5 = pd.pivot_table(df3, values=['DAILY_UNITS', 'RETAIL_PRICE', 'PROMO_PRICE', 'COMPETITOR_PRICE', 'Inventory'], index=['SALES_DATE'], columns=['Encoded_SKU_ID'], aggfunc = np.sum).fillna(0.0)
df5
```
